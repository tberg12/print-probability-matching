```python
big_letter_filelist = 'lists/faiss_image_paths_big_letters.txt'
# read in all files that have /r/ in them
with open(big_letter_filelist) as f:
    rs = [line for line in f.readlines() if '/r/' in line]
    # extract the first 4 integers before the dash in each line
    years = []
    rs_with_years = []
    for r in rs:
        year = r.split('-')[0][-4:]
        # if there's a capital letter at the end of the year
        if year[-1].isupper():
            year = r.split('-')[0][-5:-1]
        try:
            years.append(int(year))
        except ValueError:
            print(year)
            # remove r from rs
            continue
        rs_with_years.append(r)
    rs = rs_with_years

    # sort the rs by years
    rs = [r for _, r in sorted(zip(years, rs), key=lambda x: x[0])]

rs
```
