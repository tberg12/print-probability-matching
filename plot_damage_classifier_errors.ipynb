```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load the CSV
csv_path = "damage_classifier_output/3_class_data_test_damagepredictions_labels.csv"
df = pd.read_csv(csv_path)

# Compute predictions and confidence
probs = df[["normal_prob", "damage_prob", "bad_prob"]].values
df["pred"] = np.argmax(probs, axis=1)
df["pred_confidence"] = np.max(probs, axis=1)

# Identify misclassified examples
df["correct"] = df["pred"] == df["label"]
df_errors = df[~df["correct"]].copy()
df_errors["confidence_in_wrong"] = df_errors["pred_confidence"]

# Sort errors by how confident the model was in being wrong
df_errors_sorted = df_errors.sort_values(by="confidence_in_wrong", ascending=False)

# Label mapping
label_map = {0: "normal", 1: "damage", 2: "bad"}

# Display each one separately
N = 20  # Or however many you want to view
top_errors = df_errors_sorted.head(N)

for idx, row in top_errors.iterrows():
    img_path = row["path"]
    try:
        img = Image.open(img_path)
        plt.figure(figsize=(4, 4))
        plt.imshow(img, cmap='gray')
        plt.axis('off')
        plt.title(
            f"Pred: {label_map[row['pred']]} ({row['confidence_in_wrong']:.2f})\n"
            f"True: {label_map[row['label']]}"
        )
        plt.show()
    except FileNotFoundError:
        print(f"Image not found: {img_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load CSV
csv_path = "damage_classifier_output/3_class_data_test_damagepredictions_labels.csv"
df = pd.read_csv(csv_path)

# Filter to only true 'damage' label (label == 1)
damage_df = df[df["label"] == 1].copy()

# Sort by damage probability (highest to lowest)
damage_df_sorted = damage_df.sort_values(by="damage_prob", ascending=False)

# Show top N most "damaged"
N = 12  # Adjust this as needed
for idx, row in damage_df_sorted.head(N).iterrows():
    try:
        img = Image.open(row["path"])
        plt.figure(figsize=(3, 3))
        plt.imshow(img, cmap='gray')
        plt.axis('off')
        plt.title(f"Damage Score: {row['damage_prob']:.2f}")
        plt.show()
    except FileNotFoundError:
        print(f"Image not found: {row['path']}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Load the CSV
csv_path = "damage_classifier_output/3_class_data_test_damagepredictions_labels.csv"
df = pd.read_csv(csv_path)

# Get predicted and true labels
probs = df[["normal_prob", "damage_prob", "bad_prob"]].values
df["pred"] = np.argmax(probs, axis=1)
y_true = df["label"]
y_pred = df["pred"]

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Define class names
labels = ["normal", "damage", "bad"]

# Plot
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()


cm_normalized = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_normalized, annot=True, fmt=".2f", cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
