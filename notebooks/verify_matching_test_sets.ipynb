```python
chars = 'ABCDEFGHKLMNPRTW'
img_root = '/trunk2/nvog/char_images3_aligned/leviathanornaments_2022-05-15/leviathanornaments-{}-0.001-out/{}_aligned'
root = Path('/home/nvog/projects/git/anomaly-detection/matching/leviathan_matching_test_set')
chars_rows = [
    [row.strip().split(',') for row in open(root/c/'test'/'matches.csv')]
    for c in chars
]
```

```python
valid_rows = 0
invalid_rows = 0
for i, char_rows in enumerate(chars_rows):
    c = chars[i]
    print(c)
    for row in char_rows:
        if len(row) < 2:
            print('singleton row!')
            print(row)
            invalid_rows += 1
            continue
        valid_rows += 1
        annotation_utils.show_imgs_sidebyside([Image.open(Path(img_root.format(c, c))/n) for n in row], [str(n) for n in range(len(row))], figsize=(12, 4))
```
