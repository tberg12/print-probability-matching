```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import display
import os
```

```python
# Display each row of images
def display_images(row, row_index, show_col_name=False):
    """
    Displays all images in a row of the dataframe.
    :param row: A Pandas Series containing image paths for one row.
    :param row_index: The index of the current row for display.
    :param show_col_name: Whether to display column names above the images.
    """
    fig, axes = plt.subplots(1, len(row), figsize=(15, 5))
    
    if len(row) == 1:
        axes = [axes]  # To ensure axes is iterable if there's only one column

    for i, (col_name, image_path) in enumerate(row.items()):
        ax = axes[i]
        if os.path.exists(image_path):
            img = mpimg.imread(image_path)
            ax.imshow(img)
            if show_col_name:
                ax.set_title(f"{col_name}")
        else:
            ax.text(0.5, 0.5, "Image not found", ha='center', va='center', transform=ax.transAxes)
        ax.axis("off")
    
    plt.suptitle(f"Row {row_index}", fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
csv_path = "output/matching_results_nov24/output/matching_eval_nov24/_gt_test_lockespinoza_mix_neg_knn_results.csv"  # Update this to the path of your CSV file
image_data = pd.read_csv(csv_path)
for index, row in image_data.iterrows():
    display_images(row, index, show_col_name=False)
```
