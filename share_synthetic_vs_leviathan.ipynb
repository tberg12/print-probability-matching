```python
%matplotlib inline
# all_pairs.csv contains two columns with an image path per column
# load the data together and shuffle it and show the two images side by side

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image


# load the data
all_pairs = pd.read_csv('all_pairs.csv', header=None)
# all_pairs
```

```python
# show each pair of images and label each image with "Pair index <row of image>"
for index, row in all_pairs.iterrows():
    img1 = mpimg.imread(row[0])
    img2 = mpimg.imread(row[1])
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(img1)
    ax[0].set_title(str(index) + ': 1')
    ax[0].axis('off')
    ax[1].imshow(img2)
    ax[1].set_title(str(index) + ': 2')
    ax[1].axis('off')
    plt.show()
```
