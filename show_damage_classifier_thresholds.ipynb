```python
def plot_img_grid(imgs, ncols=5, figsize=(20, 20), title=''):
    nrows = int(np.ceil(len(imgs) / ncols))
    fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=figsize)
    for i, img in enumerate(imgs):
        ax = axs[i // ncols, i % ncols]
        ax.imshow(img)
        ax.axis('off')
    plt.suptitle(title, fontsize=20)
    plt.tight_layout()
    plt.show()


# plot image grid for preds_dir/damaged_{0.5, 0.6, 0.7, 0.8, 0.9}
damaged_dirs = glob.glob(os.path.join(preds_dir, "damaged_*"))
damaged_dirs
```

```python
# sample 20 images from each damaged dir
for ds in [0.5, 0.6, 0.7, 0.8, 0.9]:
    paths = random.sample(list((Path(preds_dir)/f"damaged_{ds}").glob('*')), 20)

    plot_img_grid([Image.open(p) for p in paths], title=f'damaged {ds}')
```
