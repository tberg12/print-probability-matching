```python
%matplotlib inline
# all_pairs.csv contains two columns with an image path per column
# load the data together and shuffle it and show the two images side by side

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
from matcher import get_transform

```

```python
# c=A; n=50; find /trunk2/nvog/shared/char_images3_aligned/shakespeare/shakespeare-${c}-out/synthetic/${c}_twin_samebase_noink_sz164_5_2025-03-16/train -name "*.tif" | sort | head -n ${n} | xargs -I {} imgcat {}
transform = get_transform(unpadded_image_size=164)


real_train = "/graft2/code/nvog/git/matching/damage_classifier_data/dldt_ground_truth/train_pairs.csv"
synth_train = "/graft2/code/nvog/git/matching/data/synthetic_data/train_pairs.csv"
real_train = pd.read_csv(real_train)
synth_train = pd.read_csv(synth_train)

real_train
# synth_train
```

```python
# show each pair of images and label each image with "Pair index <row of image>"
# show real_train
for index, row in real_train.iterrows():
    # load and transform images
    img1 = transform(Image.open(row['neg']))
    img2 = transform(Image.open(row['pos']))
    
    # convert tensors to numpy arrays
    img1 = img1.permute(1, 2, 0).numpy()
    img2 = img2.permute(1, 2, 0).numpy()
    
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(img1, cmap='gray')
    ax[0].set_title(f'Pair index {index}: 1')
    ax[0].axis('off')
    ax[1].imshow(img2, cmap='gray')
    ax[1].set_title(f'Pair index {index}: 2')
    ax[1].axis('off')
    plt.show()
    
    if index == 100:
        break
```

```python
from pathlib import Path
import re

char_regex = re.compile(r"([A-Za-z])_[ul]c")
split = 'train'
pando_data_dir = Path('/graft2/code/nvog/git/matching/data/synthetic_data')
# only grab columns ["file_path", "preproc_original_file_path"] and set as list of pairs
damages = [str(pando_data_dir/f'{char_regex.findall(f)[0]}_twin_samebase_noink_sz164_5_2025-03-16'/split/Path(f).name) for f in synth_train.file_path.to_list()]
normals = [str(pando_data_dir/f'{char_regex.findall(f)[0]}_twin_samebase_noink_sz164_5_2025-03-16'/split/Path(f).name) for f in synth_train.preproc_original_file_path.to_list()]

for index, (normal, damage) in enumerate(zip(normals, damages)):
    # load and transform images
    img1 = transform(Image.open(normal))
    img2 = transform(Image.open(damage))
    
    # convert tensors to numpy arrays
    img1 = img1.permute(1, 2, 0).numpy()
    img2 = img2.permute(1, 2, 0).numpy()
    
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(img1, cmap='gray')
    ax[0].set_title(f'Pair index {index}: 1')
    ax[0].axis('off')
    ax[1].imshow(img2, cmap='gray')
    ax[1].set_title(f'Pair index {index}: 2')
    ax[1].axis('off')
    plt.show()
    
    if index == 100:
        break
```
